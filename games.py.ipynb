```python
#Creating Separate Dataset from Genre Analysis based on Average User Rating.
genres_analysis=dataset[['Average User Rating','Genres']]
```

```python
genres_analysis.head(10)
```

```python
#Extracting rows with high user rating
genres_highrating =genres_analysis[genres_analysis['Average User Rating'] == 5.0] 
genres_highrating
```

```python
#List of all the Genres having 5.0 rating.
genres_unique= list(genres_highrating.Genres.unique())
genres_unique
```

```python
#Number of Genres having rating 5.0
len(genres_unique)
```

```python
#Top Genre with 5.0 rating the most
Genres = list(genres_highrating['Genres'])
counting=[]
for i in genres_unique:
    counting.append(Genres.count(i))
data={'Genres':genres_unique,'Count':counting}
genre_count = pd.DataFrame(data)
genre_count
```

```python
#Genre with most 5.0 ratings
genre_count[genre_count['Count']==max(genre_count['Count'])]
```

```python
#Plotting Genres with count of 5.0 rating .With the Plot It is also clear Games, Strategy, Simulation is the Top rated Genre
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline


plt.figure(figsize=(10,20))

plt.barh(genres_unique,counting) 
```

 Identify trend of user rating based on pricing.


```python
pricing_data =dataset[['Average User Rating','User Rating Count','Price']]
pricing_data 
```

```python
max(pricing_data['Price'])
```

```python
min(pricing_data['Price'])
```

```python
price_unique = pricing_data['Price'].unique()
price_unique
```

```python
#Extracting row for each price and performing sum on User Rating Count
total=[]
for i in price_unique:
    rows = pricing_data[pricing_data['Price']==i]
    summ = sum(rows['User Rating Count'])
    total.append(summ)
total
```

```python
data={'Price':price_unique,'Total User Rating Count':total}
pd.DataFrame(data,columns=['Price','Total User Rating Count'])
```
